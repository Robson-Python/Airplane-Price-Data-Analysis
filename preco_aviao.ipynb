## **Conjunto de dados:**

**1 - Modelo:** O modelo do avião (por exemplo, Boeing 737, Airbus A320). Representa um recurso importante que influencia o preço com base na marca e nos recursos de desempenho.

**2 - Ano de Produção:** O ano em que o avião foi fabricado. Ajuda a calcular a idade do avião, um fator significativo na depreciação.

**3 - Número de Motores:** O número total de motores no avião (1 para motores a pistão, 2 para outros).

**4 - Tipo de Motor:** Tipo de motor, como Turbofan ou Pistão. Diferentes tipos de motores afetam o desempenho, a eficiência de combustível e o preço.

**5 - Capacidade:** Capacidade de passageiros do avião. Aviões maiores com maior capacidade tendem a ter preços básicos mais altos.

**6 - Alcance em km:** Alcance máximo que o avião pode percorrer sem reabastecimento. Maior alcance aumenta a flexibilidade operacional e o custo.

**7 - Consumo de combustível em L/hora):** Consumo médio de combustível do avião. Os motores turbofan consomem menos do que os motores a pistão, influenciando os custos operacionais.

**8 - Custo de manutenção por hora:** O custo médio de manutenção por hora em USD. Custos de manutenção mais altos podem reduzir a demanda e afetar os preços.

**9 - Idade:** A idade do avião, calculada como . Aviões mais antigos tendem a ter preços mais baixos devido à depreciação. 2023 - Ano de Produção.

**10 - Região de Vendas:** A região onde o avião está sendo vendido (por exemplo, Ásia, Europa, América do Norte). A demanda regional e as variações cambiais podem afetar os preços.

**11 - Preço em USD:** O preço final do avião, servindo como **_variável_** alvo para previsão.

**Características principais**

**Realismo:** os preços são calculados dinamicamente com base em vários fatores, incluindo idade, capacidade, alcance e um fator de ruído para simular a variabilidade do mundo real.

**Diversidade:** Abrange diferentes modelos de aviões, tipos de motores e regiões para fornecer um rico conjunto de dados para diversas necessidades analíticas.

**Aplicações:** Ideal para modelagem de regressão, engenharia de recursos e análise de preços na indústria da aviação.

**Passo 3 - Correção do DataFrame (Limpeza, organização e modelagem)**

---

```python
# Substituir os valores da coluna Tipo de Motor e Região de Vendas
df_aviao["Tipo de Motor"] = df_aviao["Tipo de Motor"].replace({
    "Piston": "Pistão"
})

df_aviao["Região de Vendas"] = df_aviao["Região de Vendas"].replace({
    "Afrika": "Africa",
    "Asya": "Asia",
    "Avrupa": "Europa",
    "Avustralya": "Austrália",
    "Güney Amerika": "América do Sul",
    "Kuzey Amerika": "América do Norte"
})

display(df_aviao.head(5))

# Função para substituir os valores das colunas Modelo, Tipo de Motor e Região de Vendas de texto 
# para números e vice-versa
def codificar():
    if (df_aviao["Modelo"].dtypes == object) and (df_aviao["Tipo de Motor"].dtypes == object) and (df_aviao["Região de Vendas"].dtypes == object):
        # Codificar de texto para número
        df_aviao["Modelo"] = df_aviao["Modelo"].replace({
            "Airbus A320": 0,
            "Airbus A350": 1,
            "Boeing 737": 2,
            "Boeing 777": 3,           
            "Bombardier CRJ200": 4,  
            "Cessna 172": 5                          
        })
        
        df_aviao["Tipo de Motor"] = df_aviao["Tipo de Motor"].replace({
            "Pistão": 0,
            "Turbofan": 1
        })
        
        df_aviao["Região de Vendas"] = df_aviao["Região de Vendas"].replace({
            "Africa": 0,
            "Asia": 1,
            "Europa": 2,
            "Austrália": 3,
            "América do Sul": 4,
            "América do Norte": 5
        })
    else:
        # Cofificar de número para texto
        df_aviao["Modelo"] = df_aviao["Modelo"].replace({
            0: "Airbus A320",
            1: "Airbus A350",
            2: "Boeing 737",
            3: "Boeing 777",           
            4: "Bombardier CRJ200",  
            5: "Cessna 172"                          
        })
        
        df_aviao["Tipo de Motor"] = df_aviao["Tipo de Motor"].replace({
            0: "Pistão",
            1: "Turbofan"
        })
        
        df_aviao["Região de Vendas"] = df_aviao["Região de Vendas"].replace({
            0: "Africa",
            1: "Asia",
            2: "Europa",
            3: "Austrália",
            4: "América do Sul",
            5: "América do Norte"
        })
```

Dataframe corrigido e remodelado com valores em português.

Agora temos uma análise bem mais realista dos dados. Com a correlação ficou bem mais clara a nossa análise e agora sabemos quais as características dos aviões mais influenciam o preço. 

Caracteríticas que mais influenciam o preço segundo a correlação:

O **Alcance em KM**, a **Capacidade** e o **Tipo de Motor e seus números**: são as colunas mais próximas da Coluna Preço em USD, sendo assim, as que mais impactam no preço final do avião. 

O **Modelo** e o **Consumo de Combustível**: apesar de ter uma correlação baixa em relação ao preço, também tem relação siginificativa com o preço, sendo de grande importância no impacto do preço final do avião. 

Já a **Região das Vendas**, por ter uma correlação 0.00 em relação ao preço, possui um impacto insiginificante no preço final do avião.

Agora que sabemos quais são as colunas que influenciam o preço final, vamos agrupá-las por modelo e analisar como elas impactam no preço final. 

```python
# Agrupar a coluna Modelo com as colunas características encontradas na correlação

# Codificar colunas de número para textos
codificar()

# Agrupar pelo alcance do avião
alcance = df_aviao.groupby("Modelo")["Alcance em km"].max()
display(alcance)

# Agrupar pela capacidade do avião
capacidade = df_aviao.groupby("Modelo")["Capacidade"].max()
display(capacidade)

# Agrupar pelo número de motores do avião
numero_motor = df_aviao.groupby("Modelo")["Número de Motores"].max()
display(numero_motor)

# Agrupar pelo consumo do avião
consumo = df_aviao.groupby("Modelo")["Consumo de combustível em (L/hora)"].mean()
display(consumo)

# Agrupar pelo preço médio do avião
modelo = df_aviao.groupby("Modelo")["Preço em USD"].mean()
display(modelo)
```

Como nos mostra os agrupamentos, os modelos possuem quase as mesmas características. Os aviões mais completos, ou seja, aqueles que possuem um alcance e uma capacidade maior de passageiros, são os que justificam os preços mais elevados.

## Passos do Projeto ##

Passo 1 - Preparar o dataframe para a Inteligência Artificial

Passo 2 - Treinar a Inteligência Artificial

```python
# Importando as bibliotecas necessárias
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Criando um novo dataframe para a IA, a partir do dataframe original
df_aviao_ia = df_aviao

display(df_aviao_ia)
```

---

**Passo 1 - Preparar o dataframe para a Inteligência Artifical**

---

```python
# Codificar as colunas para tipo numérico para a Inteligência Artificial

# Codificar coluna Modelo
codificador_modelo = LabelEncoder()
df_aviao_ia["Modelo"] = codificador_modelo.fit_transform(df_aviao_ia["Modelo"])

# Codificar coluna Tipo de Motor
codificador_motor = LabelEncoder()
df_aviao_ia["Tipo de Motor"] = codificador_motor.fit_transform(df_aviao_ia["Tipo de Motor"])

# Codificar coluna Regiao de Vendas
codificador_regiao = LabelEncoder()
df_aviao_ia["Região de Vendas"] = codificador_regiao.fit_transform(df_aviao_ia["Região de Vendas"])

df_aviao_ia.info()

display(df_aviao_ia)
```

```python
# Preparar o dataframe para o treino

# y é a coluna do dataframe a ser prevista
y = df_aviao_ia["Preço em USD"]

# x são as colunas do dataframe que serão usadas para a previsão
x = df_aviao_ia.drop(columns=["Preço em USD"])

# Separar os dados de treino e os dados de teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=42)
```

---

**Passo 2 - Treinar a Inteligência Artificial**

---

```python
# Criar o medelo -> Prever os Preços em USD

# Arvore de Decisão -> RandonForest
# Nearest Neighbors -> KNN -> KNeighbors
# Regressão Linear -> RLN -> LinearRegression

# Cria a IA
modelo_arvore_decisao = RandomForestRegressor()
modelo_knn = KNeighborsRegressor()
modelo_rln = LinearRegression()

# Treinar a IA
modelo_arvore_decisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
modelo_rln.fit(x_treino, y_treino)
```

---

**Passo 3 - Escolher o melhor modelo**

---

```python
# Testar os modelos

# Testar os modelos
previsao_arvore_decisao = modelo_arvore_decisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)
previsao_rln = modelo_rln.predict(x_teste)

# Calcular os erros
print(f'MAE Árvore de Decisão: {mean_absolute_error(y_teste, previsao_arvore_decisao):,.2f}')
print(f'MAE KNN: {mean_absolute_error(y_teste, previsao_knn):,.2f}')
print(f'MAE Regressão Linear: {mean_absolute_error(y_teste, previsao_rln):,.2f}')

print(f'MSE Árvore de Decisão: {mean_squared_error(y_teste, previsao_arvore_decisao):,.2f}')
print(f'MSE KNN: {mean_squared_error(y_teste, previsao_knn):,.2f}')
print(f'MSE Regressão Linear: {mean_squared_error(y_teste, previsao_rln):,.2f}')

```

O melhor modelo é o Arvore de Decisão.

---

**Passo 4 - Usar melhor modelo**

---

```python
# Usar o melhor modelo -> Modelo Arvore de Decisão

#Importar os novos aviões para previsão
df_novos_dados = pd.read_csv("airplane_price_novos.csv")

# Codificar o novo dataframe

# Modelo
df_novos_dados["Modelo"] = codificador_modelo.fit_transform(df_novos_dados["Modelo"])

# Tipo de Motor
df_novos_dados["Tipo de Motor"] = codificador_motor.fit_transform(df_novos_dados["Tipo de Motor"])

# Regiao de Vendas
df_novos_dados["Regiao de Vendas"] = codificador_regiao.fit_transform(df_novos_dados["Regiao de Vendas"])

# Renomear as colunas do dataframe de previsão
renomear_colunas = {
    "Ano de Producao": "Ano de Produção",
    "Numero de Motores": "Número de Motores",
    "Consumo de combustivel em (L/hora)": "Consumo de combustível em (L/hora)",
    "Custo de manutencao por hora": "Custo de manutenção por hora",
    "Regiao de Vendas": "Região de Vendas",
}
df_novos_dados = df_novos_dados.rename(columns=renomear_colunas)

# Visualizar o dataframe
display(df_novos_dados)

# Fazer a previsão através do modelo árvore de decisão
nova_previsao = modelo_arvore_decisao.predict(df_novos_dados)

# Formatar os valores
nova_previsao_formatada = [f'{valor:,.2f}' for valor in nova_previsao]

display(nova_previsao_formatada)
```
